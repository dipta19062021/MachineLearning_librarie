### concat() function has obj,axis,join_axis, sort,ignore_axis,keys parameters.

```python
import pandas as pd
```

```python
sr1=pd.Series([0,1,2])
sr1
```

```python
sr2=pd.Series([3,4,5,6,7,9])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4,6,7],'Name':['A','B','C','D','E','F'], 'Class':[52,12,54,62,11,26]})
df1
```

```python
df2=pd.DataFrame({'ID':[21,22,23,24,25,27],'Name':['K','M','N','O','P','Q'], 'Class':[212,345,223,643,64,234]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis=1) #concatinate row-wise i.e after compliting one dataframe execute another   
```

```python
pd.concat([df1,df2], axis=0, ignore_index=False) #leaves the index unchanged
```

```python
pd.concat([df1,df2], axis=0, ignore_index=True) #changes the index into a contineous series of 0-n  
```

```python
pd.concat([df1,df2], axis=0, ignore_index=False)
```

```python
pd.concat([df1,df2], axis=0, ignore_index=False)
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4,6,7],'Name':['A','B','C','D','E','F'], 'Class':[52,12,54,62,11,26]})
df1
```

```python
df2=pd.DataFrame({'ID':[4,6,7],'Name':['D','E','F'], 'Class':[62,11,26]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis=1) #as in 3rd,4th,5th has no elements of Name and ID and class so NaN is being printed
```

```python
pd.concat([df1,df2], axis=1, join='inner') #only joins common rows
```

```python
pd.concat([df1,df2], axis=1, sort=False)
```

```python
pd.concat([df1,df2], keys=['First dframe','Seconf dframe'])
```

```python
pd.concat([df1,df2], axis=1, keys=['First dframe','Seconf dframe'])
```

```python
df2=pd.DataFrame({'Marks':[21,22,23,24,25,27]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis=0, sort=False)
```

```python
pd.concat([df1,df2], axis=0, sort=True)
```

```python
pd.concat([df1,df2], axis=1, sort=False)
```

```python
pd.concat([df1,df2], axis=1, sort=True)
```
