# Merge function

```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\ml\\datasets\\health care diabetes.csv')
df
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4,5,6],'class':[9.0,12.0,42.0,27.4,39.1,11.43]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4,6,7],'Name':['A','B','C','D','E','F']})
df2
```

```python
df1.merge(df2, how='inner')
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on ='ID', how='inner')
```

```python
pd.merge(df1,df2,on ='ID', how='outer')
```

```python
pd.merge(df1,df2,on ='ID', how='left', indicator=True) #indicator represent the location of present values
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8,9,2],'Name':['A','B','C','D','E','F']})
df2
```

```python
pd.merge(df1,df2, left_index=True, right_index=True) # multiindex and different values
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4,5,6],'class':[9.0,12.0,42.0,27.4,39.1,11.43]})
df2
```

```python
pd.merge(df1,df2,on='ID')# inpite of aving same values takes by name as different names are present here 
```

```python
pd.merge(df1,df2,on='ID', suffixes=('_Higher','_Middle')) #changes the suffixes 
```

```python
pd.merge(df1,df2,on='class')
```
