# use of Pivot table

# # value,fill_value,aggfunc,index,columns,margin,dropna are functions used in this pivot table

```python
df=pd.read_csv('D:\\ml\\datasets\\Book1.csv')
df
```

```python
df.pivot_table(index='Genre')#creating mean with respect to Genre , actually crestinfg mean of all the values of action, adventure and comedy
```

```python
df.pivot_table(index='Genre', columns='Year') #creating clomn of the each year and giving the mean of each row index of genre 
```

```python
df.pivot_table(index='Genre', columns='Year', aggfunc='count')
```

```python
df.pivot_table(index='Genre', columns='Year', aggfunc='sum')
```

```python
df=pd.read_csv('D:\\ml\\datasets\\Book1.csv')
df
```

```python
df.pivot_table(index='Genre', columns='Year', aggfunc='count')
```

```python
df.pivot_table(index='Genre', columns='Year', aggfunc='sum')
```

```python
df.pivot_table(index='Genre', columns='Year', fill_value=3) #fills NaN value with any value
```

```python
df.pivot_table(index='Genre', columns='Year', aggfunc='max')
```

```python
df.pivot_table(index='Genre', columns='Year',fill_value=0, margins=True) #adds up to avg of rows and coloumns
```

```python
df.pivot_table(index='Genre', columns='Year',fill_value=0, aggfunc='sum', margins=True) #adds up to sum of rows and coloumns
```
